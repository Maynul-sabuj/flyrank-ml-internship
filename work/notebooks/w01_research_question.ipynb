## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")
df = df[(df.impressions_90d > 0) & (df.content_age_days >= 90)].drop_duplicates("content_id")

n_total = len(df)
n_declining = (df.trend_direction == "down").sum()

visible = df[df.impressions_90d >= 500]
n_stale_visible = (visible.days_since_last_update >= 180).sum()

print(f"Usable pages: {n_total}")
print(f"Declining pages: {n_declining} ({n_declining/n_total:.1%})")
print(f"Visible pages: {len(visible)}")
print(f"Stale-but-visible pages: {n_stale_visible} ({n_stale_visible/len(visible):.1%})")
```
